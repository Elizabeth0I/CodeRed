```python
import pandas as pd

import os
working_directory = os.getcwd()
print(working_directory)
```

```python
path = working_directory + '/Downloads/dc_marvel_movie_performance.csv'
movie_performance = pd.read_csv(path , encoding='latin1', sep=";")
movie_performance.head()
```

```python
print(movie_performance.columns)
```

```python
# Check if the columns exist 
columns_to_remove = ['MCU', 'Phase', 'Franchise']
missing_columns = [col for col in columns_to_remove if col not in movie_performance.columns]
if missing_columns:
    print(f"Columns {missing_columns} not found in the DataFrame.")
else:
    # Remove the specified columns
    movie_performance = movie_performance.drop(columns=columns_to_remove)

    # Display the updated DataFrame after removing columns
    print("\nDataFrame after removing columns:")
    print(movie_performance.head())
```

```python
print(movie_performance.columns)
```

```python
# Get unique values of the "Character Family" column. Not sure what unique is, so going as per google suggestion = alphabetically
unique_values = sorted(movie_performance['Character Family'].unique())

# Print unique values alphabetically
print("Unique values of 'Character Family' column sorted from A to Z:")
for value in unique_values:
    print(value)
```

```python
#Longest movie time
# Find the index of the row with the Length of the films 
max_time_index = movie_performance['Length'].idxmax()

# Get the details of the longest movie
longest_movie = movie_performance.loc[max_time_index]

# Print the details of the longest movie
print("Longest Movie:")
print(longest_movie)
```

```python
# Group the DataFrame by the "Break Even" column and calculate the mean of "Rotten Tomatoes Critic Score" for each group
mean_scores = movie_performance.groupby('Break Even')['Rotten Tomatoes Critic Score'].mean()

# Print the mean scores
print("Mean values of 'Rotten Tomatoes Critic Score' per 'Break Even' category:")
print(mean_scores)
```

```python
# Define the bins and labels for the recommended categories
bins = [-float('inf'), 39, 59, 89, float('inf')]
labels = ['Strongly No', 'No', 'Yes', 'Strongly Yes']

# Add a new column "Recommended" based on the "Rotten Tomatoes Critic Score"
movie_performance['Recommended'] = pd.cut(movie_performance['Rotten Tomatoes Critic Score'], bins=bins, labels=labels)

# Print the DataFrame to show the new column
print(movie_performance.head())
```

```python
# Define the bins and labels for the recommended categories
bins = [-float('inf'), 39, 59, 89, float('inf')]
labels = ['Strongly No', 'No', 'Yes', 'Strongly Yes']

# Add a new column "Recommended" based on the "Rotten Tomatoes Critic Score"
movie_performance['Recommended'] = pd.cut(movie_performance['Rotten Tomatoes Critic Score'], bins=bins, labels=labels)

# Calculate the percentage of films falling into each recommend category
percentage = movie_performance['Recommended'].value_counts(normalize=True) * 100

# Print the percentage of films falling into recommend category
print("Percentage of films falling into each recommended category:")
print(percentage)
```

```python
# Group the data by "Character Family" and count the number of "Flops" and "Successes"
success_rate = movie_performance.groupby('Character Family')['Break Even'].value_counts().unstack().fillna(0)

# Calculate the success rate
success_rate['Success Rate'] = (success_rate['Success'] / (success_rate['Success'] + success_rate['Flop'])) * 100

# Rename the columns
success_rate = success_rate.rename(columns={'Flop': 'Number of Flops', 'Success': 'Number of Successes'})

# Reorder the columns
success_rate = success_rate[['Number of Flops', 'Number of Successes', 'Success Rate']]

# Print the success_rate DataFrame
print(success_rate)
```
